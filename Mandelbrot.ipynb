```python
import numpy as np

class Mandelbrot:
    def __init__(self, M, maxItr):
        self.M = M
        self.maxItr = maxItr
    
    def MSCalc(self, re, im): # re, im: 2d Array 
        # このサイズは適当でよい（以下でブロードキャストされるため）
        x, y = [0]*2
        
        for i in range(self.maxItr):
            x,y = x**2 - y**2 + re, 2*x*y + im
            
        return x**2 + y**2 < self.maxItr
    
M = 4
maxItr = 100

MSet = Mandelbrot(M, maxItr)

N = 100
x, y = [np.linspace(-2, 2, N)] * 2
X, Y = np.meshgrid(x, y)

plt.pcolor(X, Y, MSet.MSCalc(X, Y))
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
class Mandelbrot:
    def __init__(self, M, maxItr):
        self.M = M
        self.maxItr = maxItr
    
    def Calc(self, re, im): # re, im: 2d Array 
        x = 0.0
        y = 0.0
        
        for i in range(self.maxItr):
            if x**2 + y*2 > self.M:
                return i
            x_ = x**2 - y**2 + re
            y_ = 2*x*y + im
            x, y = x_, y_
        return self.maxItr
M = 10000000
maxItr = 100
Mand = Mandelbrot(M, maxItr)

lower = -2.5
upper = 2.5
pixel = 800
z = [[0 for m in range(pixel+1)] for n in range(pixel+1)]
for i in range(pixel+1):
    re = lower + (upper - lower)/pixel * i 
    for j in range(pixel+1):
        im = lower + (upper - lower)/pixel * j
        z[j][i] = Mand.Calc(re, im)
        
```

```python
import numpy as np
from pylab import rcParams
from pylab import *
rcParams['figure.figsize'] = 20,20

x = arange(pixel+1)
y = arange(pixel+1)
Z = array(z)

X, Y = meshgrid(x, y)
pcolor(X, Y, Z)
colorbar()
hot()
plt.gca().set_aspect('equal', adjustable='box')
show()
```
